```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import FunctionTransformer,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,mean_squared_error,accuracy_score, balanced_accuracy_score,f1_score, matthews_corrcoef
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from lazypredict.Supervised import LazyClassifier
from lightgbm import LGBMClassifier
import lightgbm

import shap
shap.initjs()
```

```python
df = pd.read_csv('/Data-path/Merged_clinicalProteinsNov29.csv', index_col = 'Final_ID')
```

## Look at meibomian gland dysfunction (MGD) for left eye since tears are from left eye

```python
df['MGD_nivå_os'] = df['MGD_nivå_os'].str.replace('Nivå ','')
#Convert the meibomian gland level to numeric
df['MGD_nivå_os']=df.MGD_nivå_os.astype('float')
```

```python
#Drop the patient with level1:
df = df[df.MGD_nivå_os!=1]
```

```python
#Remove missing values
df = df.dropna(axis=0,subset=['MGD_nivå_os'])
```

### Extract the protein quantifications

```python
only_proteins = df.iloc[:,122:]
print('There are', only_proteins.shape[1],'proteins in the dataset')
```

```python
df.MGD_nivå_os.value_counts()
```

### Trypsin, putative trypsin-6, dermacolins, dermicidins and keratines (except 18 and 19)
### should be removed from the analyses due to protein digestion/skin contamination

```python
#Import the list of proteins that should be excluded:
proteinsToRemove = pd.read_csv('contaminationProteins.csv', header=None)
```

```python
#Convert to list:
proteinList = proteinsToRemove.iloc[0,:].tolist()
```

```python
#Remove the protein columns that should not be included in the analysis
only_proteins = only_proteins.drop(columns=proteinList,axis=1)
```

```python
#Remove columns where all values for the column are NaN
only_proteins=only_proteins.dropna(axis=1, how='all')
```

```python
print('The new number of columns are:',only_proteins.shape[1])
```

## Replace all 0 values with 1 before filling in the missing values

```python
only_proteins = only_proteins.replace(0,1)
#Replace missing values with 0 (since the data is not really missing)
only_proteins=only_proteins.fillna(0)
```

## Train classifier on full dataset to predict MGD

```python
y_trainFull = df.MGD_nivå_os
```

## Log-transform and scale the protein quantifications

```python
transformer = FunctionTransformer(np.log1p, validate=True)
only_proteinsTransf = transformer.transform(only_proteins)
#Get the column names and convert back to a dataframe:
only_proteinsTransf= pd.DataFrame(only_proteinsTransf, columns = only_proteins.columns.tolist(), index=only_proteins.index)
scaler = StandardScaler()
only_proteinsTransf = scaler.fit_transform(only_proteinsTransf)
only_proteinsTransf = pd.DataFrame(only_proteinsTransf, columns = only_proteins.columns.tolist(), index=only_proteins.index)
```

### Apply Lazypredict to identify most promising classification algorithm
### https://github.com/shankarpandala/lazypredict

```python
# Must first split into training and testing:
X_train, X_test, y_train, y_test = train_test_split(only_proteins, y_trainFull,test_size=.5,random_state =123)

transformerTrain = FunctionTransformer(np.log1p, validate=True)
X_trainTransf = transformerTrain.transform(X_train)
X_testTransf = transformerTrain.transform(X_test)
#Get the column names and convert back to a dataframe:
X_trainTransf = pd.DataFrame(X_trainTransf, columns = X_train.columns.tolist(), index=X_train.index)
X_testTransf = pd.DataFrame(X_testTransf, columns = X_test.columns.tolist(),index=X_test.index)
scalerTrain = StandardScaler()
X_trainTransf = scalerTrain.fit_transform(X_trainTransf)
X_testTransf = scalerTrain.transform(X_testTransf)
```

```python
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(X_trainTransf,X_testTransf, y_train,y_test)
```

```python
models
```

## Train LGBMClassifier

```python
lgbm_class = LGBMClassifier(is_unbalance=True,num_leaves=2,random_state=42)
lgbm_class.fit(only_proteinsTransf, y_trainFull)
y_pred=lgbm_class.predict(only_proteinsTransf)

#lightgbm.plot_importance(lgbm_class,max_num_features = 15, height=3.0)
```

```python
multiclassDf = pd.DataFrame(lgbm_class.feature_name_, columns =['Features'])
multiclassDf['Importance_values']= lgbm_class.feature_importances_.tolist()
#Sort by importance
multiclassDf = multiclassDf.sort_values(by='Importance_values',ascending=True)
#Remove 0 values
multiclassDf = multiclassDf[multiclassDf['Importance_values']!=0]
#Because 4 proteins have same importance level 6, include all these in the plot
# (Or exclude all with this importance level)
multiclassDf=multiclassDf.iloc[-18:,:]
```

```python
plt.title('Feature importances for MGD prediction')
plt.xlabel('Importance value')
#plt.yticks(size=7)
plt.barh(multiclassDf['Features'],multiclassDf['Importance_values'], height=0.8)
plt.show()
```

```python
# 'Translate' the features into more understandable names...
multiclassFeaturesList = ['Adenylyl cyclase-associated protein 1','S100 calcium-binding protein A7',
                          'Immunoglobulin alpha-2 heavy chain','Proline-rich protein 4', 
                          'Immunoglobulin kappa variable 2-24','Phospholipid transfer protein','Methanethiol oxidase',
                          'X-ray repair cross-complementing protein 5',
                         'Transcobalamin-1','Heterogeneous nuclear ribonucleoprotein M','Endoplasmic reticulum resident protein 29',
                         'Phosphatidylinositol 5-phosphate 4-kinase type-2 gamma','Malate dehydrogenase, cytoplasmic',
                         'Mannose-1-phosphate guanyltransferase alpha','Prostaglandin reductase 1','Tissue alpha-L-fucosidase',
                         'Ubiquitin-conjugating enzyme E2 variant 1','Polymeric immunoglobulin receptor']
```

```python
plt.xlabel('Importance value')
plt.barh(multiclassFeaturesList,multiclassDf['Importance_values'], height=0.8)
plt.show()
```

```python
print('Accuracy:',accuracy_score(y_trainFull,y_pred))
balanced_accuracy_score(y_trainFull,y_pred)
```

```python
print('F1 score:',f1_score(y_trainFull,y_pred,average='weighted'))
print('MCC:', matthews_corrcoef(y_trainFull,y_pred))
```

```python
print(confusion_matrix(y_trainFull,y_pred))
```

```python
my_names = lgbm_class.feature_name_
```

```python
#Customized SHAP plot
#Get the global SHAP values for each feature

#Start with class 0:
shap_list0 = []
shap_list1=[]
shap_list2 = []
for i in range(2188):
    shap_list0.append(abs(shap_values[0][:,i,]).mean())
    shap_list1.append(abs(shap_values[1][:,i,]).mean())
    shap_list2.append(abs(shap_values[2][:,i,]).mean())

shap_dictFull = {}
#Add up SHAP values from all three classes (MGD levels 1-3)
for i in range(len(shap_list0)):
    shap_dictFull[my_names[i]] = shap_list0[i]
    shap_dictFull[my_names[i]] += shap_list1[i]
    shap_dictFull[my_names[i]] += shap_list2[i]
    
#Sort fom highest SHAP-value and downwards:
shap_dictFull = {k:v for k, v in sorted(shap_dictFull.items(), key = lambda item: item[1], reverse = True)}

most_imp_featuresFull = []
most_imp_valuesFull = []
for k,v in shap_dictFull.items():
    most_imp_featuresFull.append(k)
    most_imp_valuesFull.append(v)
#Choose top 18 features:
most_imp_featuresFull = most_imp_featuresFull[:18]
most_imp_valuesFull = most_imp_valuesFull[:18]


most_imp_featuresFull.reverse()
most_imp_valuesFull.reverse()
```
