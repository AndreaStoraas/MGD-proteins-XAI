```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.preprocessing import FunctionTransformer,StandardScaler
from sklearn.metrics import confusion_matrix,accuracy_score, balanced_accuracy_score,f1_score, matthews_corrcoef
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from lightgbm import LGBMClassifier
import lightgbm

import shap
shap.initjs()
```

```python
df = pd.read_csv('/Data-path/Merged_clinicalProteinsNov29.csv', index_col = 'Final_ID')
```

## Look at meibomian gland dysfunction (MGD) for left eye since tears are from left eye

```python
df['MGD_nivå_os'] = df['MGD_nivå_os'].str.replace('Nivå ','')
#Convert the meibomian gland level to numeric
df['MGD_nivå_os']=df.MGD_nivå_os.astype('float')
```

```python
#Drop the patient with level1:
df = df[df.MGD_nivå_os!=1]
```

```python
#Remove missing values
df = df.dropna(axis=0,subset=['MGD_nivå_os'])
```

### Extract the protein quantifications

```python
only_proteins = df.iloc[:,122:]
print('There are', only_proteins.shape[1],'proteins in the dataset')
```

### Trypsin, putative trypsin-6, dermacolins, dermicidins and keratines (except 18 and 19)
### Should be removed from the analyses due to cleavage/skin contamination

```python
#Import the list of proteins that should be excluded:
proteinsToRemove = pd.read_csv('contaminationProteins.csv', header=None)
```

```python
#Convert to list:
proteinList = proteinsToRemove.iloc[0,:].tolist()
```

```python
#Remove the protein columns that should not be included in the analysis
only_proteins = only_proteins.drop(columns=proteinList,axis=1)
```

```python
#Remove columns where all values for the column are NaN
only_proteins=only_proteins.dropna(axis=1, how='all')
```

```python
print('The new number of columns are:',only_proteins.shape[1])
```

## Replace all 0 values with 1 before filling in the missing values

```python
only_proteins = only_proteins.replace(0,1)
#Replace missing values with 0 (since the data is not really missing)
only_proteins=only_proteins.fillna(0)
```

## Train classifier on full dataset to predict MGD

```python
y_trainFull = df.MGD_nivå_os
```

## Log-transform and scale the protein quantifications

```python
transformer = FunctionTransformer(np.log1p, validate=True)
only_proteinsTransf = transformer.transform(only_proteins)
#Get the column names and convert back to a dataframe:
only_proteinsTransf= pd.DataFrame(only_proteinsTransf, columns = only_proteins.columns.tolist(), index=only_proteins.index)
scaler = StandardScaler()
only_proteinsTransf = scaler.fit_transform(only_proteinsTransf)
only_proteinsTransf = pd.DataFrame(only_proteinsTransf, columns = only_proteins.columns.tolist(), index=only_proteins.index)
```

## Train LGBMClassifier

```python
lgbm_class = LGBMClassifier(is_unbalance=True,num_leaves=2,random_state=42)
lgbm_class.fit(only_proteinsTransf, y_trainFull)
y_pred=lgbm_class.predict(only_proteinsTransf)
#Also get the predicted probabilities
y_predProba = lgbm_class.predict_proba(only_proteinsTransf)
```

### Inspect predicted probabilities

```python
#Create a df with the predicted probabilities for each MGD level:
probaDF = pd.DataFrame(y_predProba, columns =['Level 2','Level 3','Level 4'])
probaDF['PredictedLevel'] = y_pred
probaDF['TrueLevel'] = y_trainFull.tolist()
probaDF.index = y_trainFull.index
probaDF['UncertaintyGroup'] = 'Nan'
```

### If ground truth is known, we up-weight SHAP-values for correct predictions with high certainty and down-weight SHAP-values for predictions with low certainty. Borderline SHAP-values are unchanged

```python
negativeSHAP2 = pd.DataFrame(columns = my_names)
```

```python
#Get the SHAP values for the negative group:
for j in range(probaDF2.shape[0]):
    #Get the negative=incorrect observations:
    if probaDF2.iloc[j,-1]=='Negative':
        #Get the predicted class and probability:
        predictedClass = probaDF2.iloc[j,1]
        predictedProba = probaDF2.iloc[j,0]
        #Get the corresponding SHAP values
        SHAP_values = shap_valuesLevel2[j]
        #Reduce the SHAP value since wrong prediction (multiplication since probability never exceeds 1)
        #Also want more reduction to the most certain predicted probabilities (1 - proba)
        if predictedClass == 1:
            correctedSHAP_valuesNegative = SHAP_values*(1-predictedProba)
        #If predicted class is 0, we need to flip the probability
        else:
            flippedProba = 1-predictedProba
            correctedSHAP_valuesNegative = SHAP_values*(1-flippedProba)
        newDF = pd.DataFrame(correctedSHAP_valuesNegative).T
        newDF.columns = my_names
        negativeSHAP2 = pd.concat([negativeSHAP2,newDF],axis=0)
```

```python
positiveSHAP2 = pd.DataFrame(columns = my_names)
```

```python
#Get the SHAP values for the positive group:
for j in range(probaDF2.shape[0]):
    #Get the positive=correct observations:
    if probaDF2.iloc[j,-1]=='Positive':
        #Get the predicted probability:
        predictedProba = probaDF2.iloc[j,0]
        #Get the corresponding SHAP values
        SHAP_values = shap_valuesLevel2[j]
        #Increase the SHAP value since right prediction (division since probability never exceeds 1)
        #Also want more increase for the most certain predicted probabilities (1 - proba)
        if predictedClass == 1:
            correctedSHAP_valuesPositive = SHAP_values/(1-predictedProba)
        #If predicted class is 0, we need to flip the probability
        else:
            flippedProba = 1-predictedProba
            correctedSHAP_valuesPositive = SHAP_values/(1-flippedProba)
        newDF = pd.DataFrame(correctedSHAP_valuesPositive).T
        newDF.columns = my_names
        positiveSHAP2 = pd.concat([positiveSHAP2,newDF],axis=0)
```

```python
borderlineSHAP2 = pd.DataFrame(columns = my_names)
```

```python
#Get the SHAP values for the borderline group:
for j in range(probaDF2.shape[0]):
    #Get the positive=correct observations:
    if probaDF2.iloc[j,-1]=='Borderline':
        #Get the predicted probability:
        predictedProba = probaDF2.iloc[j,0]
        #Get the corresponding SHAP values (which we don't change)
        SHAP_values = shap_valuesLevel2[j]
        newDF = pd.DataFrame(SHAP_values).T
        newDF.columns = my_names
        borderlineSHAP2 = pd.concat([borderlineSHAP2,newDF],axis=0)
```
